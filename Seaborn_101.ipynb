```python
# set up the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
#laod data ( built in data)
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

# **Distribution Plot**

**Distribution plot**

```python
#sns.distplot(crash_df['not_distracted'] , kde=False ,bins = 25)  # to not show the line
sns.distplot(crash_df['not_distracted'] , bins = 25)
```

**Joint Plot**

```python
sns.jointplot(x='speeding' , y='alcohol' , data= crash_df , kind='reg')
```

**KDE Plot**

```python
sns.kdeplot(crash_df['alcohol'])
```

**Pair Plots:**

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df , hue='sex' , palette='Blues')
```

**Rug plot**

```python
sns.rugplot(tips_df['tip'])
```

**Styling**

```python
sns.set_style('darkgrid')
sns.set_context('talk', font_scale= 1)

sns.jointplot(x='speeding' , y='alcohol' , data= crash_df , kind='reg')
sns.despine(left = True , bottom= True)

plt.figure(figsize=(5,8))
```

# **Categorical plots**

**Bar plots**

```python
sns.barplot(x='sex' , y='total_bill',  data=tips_df , estimator = np.mean)
```

**Count plot**

```python
sns.countplot(x='sex', data =tips_df)
```

**Box plot**

```python
sns.boxplot(x='day' , y='total_bill', data=tips_df , hue='sex')
plt.legend(loc=0)
```

**Violin Plot**

```python
sns.violinplot(x='day' , y='total_bill', data=tips_df , hue='sex' , split=True)
```

**Strip Plot**

```python
sns.stripplot(x='day' , y='total_bill', data=tips_df , jitter = True , hue='sex')
   # we can use dodge=True to seperate the males from females
```

**Swarm plot**                                                                                           

```python
sns.swarmplot(x='day' , y='total_bill', data=tips_df)
```

**Palettes**

```python
plt.figure(figsize=(5,8))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day' , y='total_bill', data=tips_df , hue='sex', palette='magma')
plt.legend(loc=0)
#loc=1 :upper right , loc=2 : upper left , loc=3 : lower right , loc=4 : lower left
```

# **Matrix plots:**

**heat map**

```python
plt.figure(figsize=(8,6))
sns.set_context('paper' , font_scale=1.4)

crash_mx = crash_df.corr()

sns.heatmap(crash_mx , annot=True , cmap='Blues')
```

```python
flights_df = sns.load_dataset('flights')
flights = flights_df.pivot_table(index='month' , columns='year' , values='passengers' )
sns.heatmap(flights , cmap='Blues' , linecolor='white' , linewidth=2)
```

**Cluster map**

```python
iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
```

```python
sns.clustermap(flights , cmap='Blues' , standard_scale=1)
#The clustermap reposition data to find clusters of data
```

**Pair Grid**

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris , hue="species")
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g.map_diag(plt.hist)
```

**Facet grid**

```python
tips_fg = sns.FacetGrid(tips_df , col='time' , row='smoker')
tips_fg.map(plt.hist , 'total_bill' , bins=8)
```

```python
kws = dict(s=50 , linewidth=.5 , edgecolor='w')
tips_dfm  = sns.FacetGrid(tips_df , col='sex' , hue='smoker' , height=4 , aspect = 1.3, hue_order=['Yes' , 'No'],
                          hue_kws=dict(marker=['^' , 'v']))
tips_dfm.map(plt.scatter , 'total_bill' , 'tip' , **kws)
```
